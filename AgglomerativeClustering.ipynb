```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Mall_customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:,3:5].values
```

```python
X
```

```python

from scipy.cluster.hierarchy import dendrogram , linkage
dendrogram1 = dendrogram(linkage(X,method ='ward'))
plt.title('Dendogram')
plt.xlabel('customers')
plt.ylabel('euclidean distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=5,  affinity='euclidean', linkage='ward')
y_pred = model.fit_predict(X)
```

```python
y_pred
```

```python
plt.scatter(X[y_pred == 0, 0], X[y_pred == 0, 1], s=100, c="red", label="cluster1")
plt.scatter(X[y_pred == 1, 0], X[y_pred == 1, 1], s=100, c="yellow", label="cluster2")
plt.scatter(X[y_pred == 2, 0], X[y_pred == 2, 1], s=100, c="blue", label="cluster3")
plt.scatter(X[y_pred == 3, 0], X[y_pred == 3, 1], s=100, c="black", label="cluster4")
plt.scatter(X[y_pred == 4, 0], X[y_pred == 4, 1], s=100, c="cyan", label="cluster5")

plt.title("AgglomerativeClustering")
plt.xlabel("Customers")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```
